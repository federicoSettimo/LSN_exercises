### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

filein = open("1/out1.txt") # reading data from the output obtained with C++
M = int(filein.readline())
N = int(filein.readline())
L = int(M/N)
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
x = np.arange(N)
x*=L
for i in range(N):
    sum_prog[i] = float(filein.readline())
for i in range(N):
    err_prog[i] = float(filein.readline())
filein.close()

z = np.zeros(N)
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.plot(x,z)
plt.xlabel('#throws')
plt.ylabel(r'$<I>-1$')
plt.grid(True)
plt.show()
```

### Exercise 02.2
- <span style="color:red">3D Random Walks (RW) on a cubic lattice and in the continuum:</span> Repeat many times (e.g. say $10^4$) the simulation of a random walk in 3D always starting at the origin

1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$

```python
# to do
# - write functions to calculate errors and sqrts?
# - print all errors or only some of them?
# - error with blocks

filein = open("2/out1.txt")
N = int(filein.readline())
r = np.zeros(N)
err = np.zeros(N)
for i in range(N):
    r[i] = float(filein.readline())
# printing error only for odd values of i (otherwise the plot is too "crowded")
for i in range(N):
    if i%2 == 1:
        err[i] = float(filein.readline())
filein.close()

x = np.arange(N)
x += 1
plt.errorbar(x,r,yerr=err)
plt.xlabel(r'$i$')
plt.ylabel(r'$\sqrt{<|r|^2>}$')
plt.grid(True)
plt.show()
```

```python
filein = open("2/out2.txt")
N = int(filein.readline())
for i in range(N):
    r[i] = float(filein.readline())
# printing error only for odd values of i (otherwise the plot is too "crowded")
for i in range(N):
    if i%2 == 1:
        err[i] = float(filein.readline())
filein.close()

plt.errorbar(x,r,yerr=err)
plt.xlabel(r'$i$')
plt.ylabel(r'$\sqrt{<|r|^2>}$')
plt.grid(True)
plt.show()
```
